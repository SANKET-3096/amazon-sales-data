```python
# importing required library 
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# 
df = pd.read_csv(r"C:\Users\plava\Downloads\data.csv")
df
```

```python
data = {'order_id': np.arange(1000),
        'order_date': pd.date_range(start='2023-01-01', periods=1000),
        'product_id': np.random.randint(100, size=1000),
        'quantity': np.random.randint(1, 10, size=1000),
        'price': np.random.uniform(10, 100, size=1000)}

df = pd.DataFrame(data)
```

```python
df['total_amount'] = df['quantity'] * df['price']
df['order_month'] = df['order_date'].dt.month
df['order_year'] = df['order_date'].dt.year
```

```python
monthly_sales = df.groupby('order_month')['total_amount'].sum()
monthly_sales.plot(kind='bar', title='Monthly Sales')
plt.show()


```

```python
yearly_sales = df.groupby('order_year')['total_amount'].sum()
yearly_sales.plot(kind='bar', title='Yearly Sales')
plt.show()
# Group by year and month and sum quantities
```

```python
yearly_monthly_sales = df.groupby(['order_year', 'order_month'])['total_amount'].sum()
yearly_monthly_sales.plot(kind='bar', title='Yearly_month Sales')
plt.show()
```

```python

# Ensure all lists have the same length (example: duplicate dates)
num_days = len(pd.date_range(start='2023-01-01', end='2024-12-31', freq='D'))
dates = pd.date_range(start='2023-01-01', end='2024-12-31', freq='D')[:num_days]
products = np.random.choice(['A', 'B', 'C'], size=num_days)
quantities = np.random.randint(1, 10, size=num_days)

data = {'date': dates, 'product': products, 'quantity': quantities}
df = pd.DataFrame(data)
```

```python
df['month'] = df['date'].dt.month
month_wise_data = df.groupby('month')['quantity'].sum().reset_index()
```

```python
month_wise_data['total_quantity'] = month_wise_data['quantity'].sum()
month_wise_data['average_quantity'] = month_wise_data['quantity'].mean()
```

```python
month_wise_data.plot(x='month', y='quantity')
plt.show()
```
